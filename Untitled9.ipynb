```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np
```

```python
img = cv2.imread('/content/DSCN0479-001.JPG')
cv2_imshow(img)
```

```python
def filter_img(image):
  identity = cv2.bilateralFilter(image, 5, 20, 30)
  cv2_imshow(identity)
filter_img(img)  
```

```python
image = img.astype(np.float32)
shape = img.shape[:2]

mean = 0
sigma = 0.005 ** 0.5  # var*0.5
gamma = 0.25
alpha = 0.75
beta = 1 - alpha

gaussian = np.random.normal(loc = mean, scale = sigma, size = (shape[0], shape[1], 1)).astype(np.float32)
gaussian = np.concatenate((gaussian, gaussian, gaussian), axis = 2)
gaussian_img = cv2.addWeighted(image, alpha, beta*gaussian, beta, gamma)

cv2_imshow(gaussian_img)
```

```python
Box_Filter = cv2.boxFilter(img, -1, (7, 7), False)
cv2_imshow(Box_Filter)
```

```python
Median_Blur = cv2.medianBlur(img, 7)
cv2_imshow(Median_Blur)
```

```python
error1 = np.sum((img.astype("float") - Median_Blur.astype("float")) ** 2)
error1 /= float(img.shape[0] * img.shape[1])

error2 = np.sum((img.astype("float") - Box_Filter.astype("float")) ** 2)
error2 /= float(img.shape[0] * img.shape[1])

error3 = np.sum((img.astype("float") - gaussian_img.astype("float")) ** 2)
error3 /= float(img.shape[0] * img.shape[1])

print(error1)
print(error2)
print(error3)
```

```python
err1 = np.square(np.subtract(img,Box_Filter)).mean()
err2 = np.square(np.subtract(img,Median_Blur)).mean()
err3 = np.square(np.subtract(img,gaussian_img)).mean()

print(err1)
print(err2)
print(err3)
```

```python
max_pixel = 255.0

psnr1 = 20 * np.log10(max_pixel / np.sqrt(err1))
psnr2 = 20 * np.log10(max_pixel / np.sqrt(err2))
psnr3 = 20 * np.log10(max_pixel / np.sqrt(err3))

print(psnr1)
print(psnr2)
print(psnr3)
```
